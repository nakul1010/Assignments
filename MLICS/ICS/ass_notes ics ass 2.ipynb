**  Assignment 2 AES**

<br>
Advanced Encryption Standard (AES) is a specification for the encryption of electronic data established by the U.S National Institute of Standards and Technology (NIST) in 2001. AES is widely used today as it is a much stronger than DES.<br>
Points to remember<br>
•	AES is a block cipher.<br>
•	The key size can be 128/192/256 bits.<br>
•	Encrypts data in blocks of 128 bits each.<br>
<br>

The number of rounds depends on the key length as follows :<br>
•	128 bit key – 10 rounds<br>
•	192 bit key – 12 rounds<br>
•	256 bit key – 14 rounds<br>

aes.png


Encryption round comprises of 4 steps :<br>
•	SubBytes :   This step implements the substitution. In this step each byte is substituted by another byte. Its performed using a lookup table also called the S-box. This substitution is done in a way that a byte is never substituted by itself. The result of this step is a 16 byte (4 x 4 ) matrix like before.
<br>
•	ShiftRows :  Each row is shifted a particular number of times.<br>
•	The first row is not shifted<br>
•	The second row is shifted once to the left.<br>
•	The third row is shifted twice to the left.<br>
•	The fourth row is shifted thrice to the left.<br>

<br>
•	MixColumns :  This step is basically a matrix multiplication. Each column is multiplied with a specific matrix and thus the position of each byte in the column is changed as a result. Here each word (4 x 1 bytes a column) is multiplied with  matrix(4 x 4) and resultant (4x1) is generated.<br>
    This step is skipped in the last round.
<br>

Add Round Key : Now the resultant output of the previous stage is XOR-ed with the corresponding round key.<br>

Main note : In this algorithm all the bits are considered in 128 bits block and while performing algorithm , 128 bits are converted into (4x4  16 bytes) matrix.
<br>
Decryption :<br>
The stages in the rounds can be easily undone as these stages have an opposite to it which when performed reverts the changes. Each 128 blocks goes through the 10,12 or 14 rounds depending on the key size.<br>
The stages of each round in decryption is as follows :<br>
•	Add round key<br>
•	Inverse MixColumns :  This step is similar to the MixColumns step in encryption, but differs in the matrix used to carry out the operation.<br>
•	ShiftRows<br>
•	Inverse SubByte : Inverse S-box is used as a lookup table and using which the bytes are substituted during decryption.<br>

Even though its been 20 years since its introduction we have failed to break the AES algorithm as it is infeasible even with the current technology. Till date the only vulnerability remains in the implementation of the algorithm.

Diffie Hellman (DH) key exchange algorithm is a method for securely exchanging cryptographic keys over a public communications channel. It is an asymmetric key exchange algorithm which uses both public and private key . The protocol is secure only if the authenticity of the 2 participants can be established. There are 2 publicly known numbers :A prime number p and an integer root that is a primitive root of prime.<br>
<br>
Note  : A root is called a primitive root of prime if all the powers ranging between (1 , prime-1)  and the numbers generated by raising primitive root to all those powers are unique and lie in range (1 , prime_no -1).<br>
Eg : 3 is primitive root of  prime no = 7.<br>
31 mod 7 = 3<br>
32 mod 7 = 2<br>
33 mod 7 = 6<br>
34 mod 7 = 4<br>
35 mod 7 = 5<br>
36 mod 7 = 1<br>
All numbers are unique and within range (1 , 6).<br>

Public_key_a = (  (primitive_root)private_key_a ) % (prime_no)<br>

K-Means clusterings is an unsupervised learning algorithm which do not contain any dependent variable and the end goal is to assign each point to a cluster . Firstly any k -random points are selected as clusters  and then following steps are followed :
1.	Calculate the distance of a point with every centroid and assign the point to the closest centriod.
2.	After assigning the point to a centriod , re-calculate that centriod by taking the avg. of both the points.
3.	Repeat  steps 1 and 2 for all the data points.
4.	Finally re-calculate all the centriods by taking the avg. of points in their respective centroid. 
5.	Repeat above process till all points are not settled in a cluster.
